```python
# generate data of 2 independent variables (X) and 1 dependent variable (y) with a linear relationship and a bit of noise

import numpy as np
import matplotlib.pyplot as plt

# generate 3 2D vectors, with elements each between 0 and 1
X = np.random.rand(3, 2)

y = 2*X[:,0] + 3*X[:,1] + np.random.rand(3)

# plot the data as a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], y, c='r', marker='o')
ax.set_xlabel('X0 Label')
ax.set_ylabel('X1 Label')
ax.set_zlabel('y Label')
plt.show()


```

```python
print(f'X={X}')
print(f'y={y}')
```

```python
print(y.reshape(-1,1))
```
